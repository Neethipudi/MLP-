# PROGRAM 4: Write a program to implement different normalization techniques using Python.
(a) min-max normalization by setting min = 0 and max = 1
(b) z-score normalization
(c) z-score normalization using the mean absolute deviation
(d) normalization by decimal scaling

# a) Positive numbers only

```python
# Importing necessary libraries
import statistics
import pandas as pd

# Define Min-max normalization
def minMaxNor(num,list):
    minNum=int(input("Enter New Minimum Value:"))
    maxNum = int(input("Enter New Maximum Value:"))
    ans=round(((num-min(list))/(max(list)-min(list))*(maxNum-minNum))+minNum,2)
    return ans

# Define Z-score normalization using Mean Absolute Deviation (MAD)
def zNorMAD (num,mean,abMeanDiv):
    return round((num-mean)/abMeanDiv,2)

# Define Decimal Scaling normalization
def decNor(num,maxNum):
    digit=len(str(maxNum))
    div=pow(10,digit)
    return num/div

# Define positive numbers
data=[200, 300, 400, 600, 1000]

# Normalization
num=int(input("\nEnter an item from data :"))
print("\nAfter doing min-max normalization :",minMaxNor(num,data))
print("\nAfter doing z-score normalization using SD:", zNor(num,statistics.mean(data),statistics.stdev(data)))
df = pd.DataFrame(data)
print("\nAfter doing z-score normalization using MAD:", zNorMAD(num,statistics.mean(data),df.mad()))
print("\nAfter doing decimal scaling normalization :", decNor(num,max(data)))
```
