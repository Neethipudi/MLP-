```
#import the numpy library and the load_wine dataset for classification

import numpy as np
from sklearn.datasets import load_wine
```

```
#Load and analyze the dataset

wine_dataset = load_wine()

# print(wine_dataset)
wine_dataset_features = wine_dataset.data


print("Wine dataset feature values are: \n", wine_dataset_features,"\n")
print("Wine dataset labels are: \n", wine_dataset_labels, "\n")
```

```
#Split the dataset into training and testing sets

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test.split(wine_dataset_features, wine_dataset_labels, test_size=0.20, random_state=28)
```

```
#Analyze the training and testing sets

print("Training features is: \n", x_train, "\n")
print("Training labels are: \n", y_train, "\n")
print("Testing labels are: \n", y_test, "\n")

print("Shape of training data is: ", np.shape(x_train), "\n")
print("Shape of training labels is: ", np.shape(y_train), "\n")
print("Shape of testing dataset is: ", np.shape(x_test), "\n")
print("Shape of testing labels is: ", np.shape(y_test), "\n")
```

```
# Loading the K-Nearest Neighbor classifier

from sklearn.neighbors import KNeighborsClassifier

knn_classifier = KNeighborsClassifier(n_neighbors=5)
print(knn_classifier)
```

```
#Fitting the classifier on the training dataset

knn_classifier.fit(x_train,y_train)
```

```
print(np.shape(x_test))
```

```
#Making predictions with KNN classifier, comparing it with ground-truth and getting the score value on the test dataset
knn_classifier_predictions = knn_classifier.predict(x_test)
print("KNN classifier predictions on the test data are: \n", knn_classifier_predictions, "\n")
print(np.shape(knn_classifier_predictions))

print("Comparing the ground truth with the predictions [ground-truth  knn-predictions] \n", np.column_stack((y_test,knn_classifier_predictions)), "\n")
print("KNN classifier score on the test dataset is: ", knn_classifier.score(x_test,y_test), "\n")
```
